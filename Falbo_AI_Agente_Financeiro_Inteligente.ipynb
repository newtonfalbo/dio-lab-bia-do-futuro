# Falbo AI - Agente Financeiro Inteligente

Projeto desenvolvido para o laboratório **BIA do Futuro** da DIO.

## Objetivos

- Consultar o perfil do investidor;
- Recuperar histórico de atendimentos;
- Recomendar produtos financeiros;
- Analisar transações financeiras;
- Demonstrar o uso de IA Generativa com base de conhecimento estruturada.

Bibliotecas

```python
import pandas as pd
import json
from pathlib import Path
```

Carregamento dos Dados

```python
import os

for raiz, diretorios, arquivos in os.walk("."):
    for arquivo in arquivos:
        print(os.path.join(raiz, arquivo))
```

```python
DATA_PATH = Path("data")

historico = pd.read_csv(DATA_PATH / "historico_atendimento.csv")
transacoes = pd.read_csv(DATA_PATH / "transacoes.csv")

with open(DATA_PATH / "perfil_investidor.json", "r", encoding="utf-8") as f:
    perfil = json.load(f)

with open(DATA_PATH / "produtos_financeiros.json", "r", encoding="utf-8") as f:
    produtos = json.load(f)

print("Dados carregados com sucesso!")
```

Visualização da Base

```python
print("Perfil do Investidor")
```

Consulta do Perfil

```python
def consultar_perfil():
    return perfil

consultar_perfil()
```

Recomendação de Produtos

```python
def recomendar_produtos():

    perfil_cliente = perfil["perfil"]

    recomendados = []

    for produto in produtos:
        if perfil_cliente.lower() in produto["indicado_para"].lower():
            recomendados.append(produto)

    return recomendados

recomendar_produtos()
```

Último Atendimento

```python
def ultimo_atendimento():
    return historico.tail(1)

ultimo_atendimento()
```

Análise de Transações

```python
def analisar_transacoes():

    media = transacoes["valor"].mean()

    suspeitas = transacoes[
        transacoes["valor"] > media * 3
    ]

    return suspeitas

analisar_transacoes()
```

Simulação do Agente

```python
def falbo_ai(pergunta):

    pergunta = pergunta.lower()

    if "perfil" in pergunta:
        return consultar_perfil()

    elif "produto" in pergunta or "investimento" in pergunta:
        return recomendar_produtos()

    elif "atendimento" in pergunta:
        return ultimo_atendimento()

    elif "transação" in pergunta:
        return analisar_transacoes()

    else:
        return "Não encontrei informações na base de conhecimento."
```

Exemplos de Uso

```python
falbo_ai("Qual é o meu perfil?")
```

```python
falbo_ai("Qual investimento é indicado para mim?")
```

```python
falbo_ai("Qual foi meu último atendimento?")
```

```python
falbo_ai("Existe alguma transação suspeita?")
```

O Falbo AI demonstra como um agente financeiro pode utilizar uma base de conhecimento estruturada para fornecer respostas contextualizadas, personalizadas e confiáveis.

A solução integra informações de perfil do investidor, histórico de atendimentos, catálogo de produtos financeiros e histórico de transações para apoiar a tomada de decisão financeira.
